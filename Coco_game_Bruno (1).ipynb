```python
#Start by defining the first function to store the name of the player
def get_name():
    name = input(prompt = "Hi, what is your name? ")
    return name

#Print a welcome message with the name of the player
player_name = get_name()
print(f"Hi {player_name}, welcome to this adventure game!") 

#Add separators in every stage
input("Press any key to continue\n")

#Provide context of the game to the player 
print("You're in the exciting game of Coco, a wonderful pixar movie based on the mexican tradition of El día de los Muertos or the Day of the Dead")
input(prompt = "Press any key to continue\n")

print("--" * 55)

#Provide context of the story to the player
print("Meet Miguel, a mexican boy who dreams of being a musician like his great idol, Ernesto de la Cruz, the most famous singer of all time; however, Miguel's family is against him pursuing the dream of being a musician since Mamá Coco, Miguel's great-grandmother, was abandoned by his father, who was a musician and singer with unknown identity, because of this, Miguel's family wants him to forget about music and join the family tradition of making and embroidering shoes, but Miguel wants to fight for his great dream and follow in the footsteps of his idol, he suspects that his great-great-grandfather is Ernesto de la Cruz and is willing to do everything to find it out, but the only way to do it is visiting the land of the dead.")
input(prompt = "Press any key to read the instructions\n")

print("--" * 55)

#Provide instructions to the player
print(f"Your mission is to help Miguel get into the land of the dead, discover if his idol, Ernesto de la Cruz, is his real great-great-grandfather and then help him come back to the land of the living people, to win the game, you will have to pass all the levels in less than 3 minutes by answering different questions and solving puzzles correctly.")
input(f"Press any key to start playing, your time starts now. Good luck, {player_name}!")

print("--" * 55)


#Import time to start measuring time
import time
start_time = time.time()
    


import random

#A def function is created to store all the code in the first stage
def stage_1():
    

    
    #The instructions are printed out and the input is showed to the player to start the game
    print(f"Welcome to level 1, you need at least 2 correct answers to help Miguel enter to the Land of the Dead, answer correctly the following questions:")
    input(prompt = "Press enter for question 1\n")
    
    #A list with all the questions is created
    questions = ["What is Día de los Muertos or the Day of the Dead?",
                 "When is the Day of the Dead celebrated?", 
                 "What is an ofrenda?", 
                 "What is the belief behind the day of the dead?"]
    
    #A second list with all the options is created
    options = [("A. A harvest festival", "B. A festival of lights", "C. A day to honor and remember the deceased", "D. A celebartion of winter"),
               ("A. November 2", "B. November 5", "C. October 31", "D. May 5"),
               ("A) A famous song", "B) A type of mexican dance", "C) An altar to remember departed people", "D) A classic mexican dish"),
               ("A. Dead people temporarily return to the world of the living.", "B. All the dead people are resurrected", "C. Predcit the future with spiritual techniques", "D. The same as Halloween")]
    
    #Let's create another list with the correct answers
    answers = ["C", "A", "C", "A"]
    
    guesses = []
    points = 0
    
    #Add a variable named question_num to gradually increase the index from the options list
    question_num = 0
    
    #Insert a for loop to evaluate the questions list and show each question
    for question in questions:
        print("--" * 55)
        print(question)
        
        #Insert another for loop to evaluate the options list starting from index 0
        for option in options[question_num]:
            print(option)
        
        #Insert the input to get the answer from the player 
        choice = input(prompt = "Enter your answer: ").upper()
        
        #Append the answer from the player to the guesses list
        guesses.append(choice)
        
        #Insert a conditional statement to see if the iput matches the answer
        if choice == answers[question_num]:
            
            #In case the answer is correct, one point is added, one point is one correct answer
            points += 1
            #In case the answer is correct, the following messages are printed out
            print("--" * 55)
            print("That's correct!")
        
        #In case the answer is incorrect, the player is informed and the correct answer is showed
        else:
            print("--" * 55)
            print("That is incorrect!")
            print(f"The correct answer is {answers[question_num]}")
        
        #Increase by 1 the index from questions and answers
        question_num += 1
    
    print("--------------------------------------------------------------------------")
    
    #If the player responds correctly 3 or more questions, print a Well done message and level 2 begins
    if points >= 3:
        print("Well done! You helped Miguel to enter the land of the dead")
        stage_2()
    
    #If not, the player can answer an extra_question to recover some points
    else:
        extra_question()
        
#Insert a def function to store the extra question
def extra_question():
    
    #Start by printing out a message to provide context
    print("You didn't answer correctly to pass to the next level, but you still have one opportunity! Let's try again with another question")
    
    
    input(prompt = "Press enter to see the question")
    
    #Define a variable with the extra question
    extra_question = "In the film, how does Miguel access to the Land of the Dead?"
    
    #Define a list with the posible options
    extra_options = [("A. Drinking a magic potion",  
                      "B. Crossing a bridge of flowers", 
                      "C. Playing a special guitar", 
                      "D. Following a dog named Dante")]
    
    #Store the correct answer with the variable extra_answer
    extra_answer = "C"
    
    print("--" * 55)
    
    #Print out the variable with the question
    print(extra_question)
    
    #Provide a hint to the player 
    print("Hint: Ernesto de la Cruz was always seen with one in hand")
    
    #define a for loop to evaluate and print out the extra options list
    for extra_option in extra_options:
            print(extra_options)
            
            #Ask the player to insert the input
            extra_choice = input(prompt = "Enter your answer: ").upper()
            
            #Conditional statement to verify if the input matches with the correct answer
            if extra_choice == extra_answer:
                print("--" * 55)
                print("Nice job! Miguel is now in the Land of the Dead")
                stage_2()
                
            #If the answer is wrong, the player still playing with less time
            else: 
                print("--" * 55)
                print(f"Wrong, the correct answer is {extra_answer} but don't worry! You can continue to the next level with less time, hurry up!")
                stage_2()
             
    
#Define the function stage_2 for the second level
def stage_2():
    
    
    input(prompt = "Press enter to start level 2")
    print("--" * 55)
    
    #Begining the level by explaining the context of the movie
    print("Now Miguel is inside the Land of the Dead. In this magical and brilliant world, all the deceased people rest in peace. They are remembered year after year by their loved ones from the land of the living and can cross the bridge that unites both worlds when their relatives put their photos in the ofrendas accompanied by candles and dishes that they enjoyed when they were alive. At this point, Miguel finds his loved ones who have passed away and begins the adventure to find his idol. Your mission is to help him find the great Ernesto de la Cruz and find out if he is the father of Mama Coco; for this, you must guide Miguel and solve a riddle about the tradition of the Day of the Dead to pass to the next level.")
    
    input(prompt = "Press enter to get the first question")
    
    
    print("--" * 55)
    
    #Now store the questions in 3 different variables 
    quest_1 = "Which  iconic character from the Land of the Dead do you think can help Miguel find Ernesto de la Cruz?"
    quest_2 = "Where do you think Miguel can find Ernesto de la Cruz?"
    quest_3 = "How can Miguel get Ernesto's attention?"
    
    #Store the options for each question in 3 different lists
    options_1 = ["A. Hector, an unknown deceased", "B. Dante, Miguel's dog", "C. Mama Imelda, Coco's mother", "D. Chicharron, a former musician"]
    options_2 = ["A. In the land of the living", "B. On the bridge that connects both worlds", "C.  In the City of the Dead", "D. At the bar"]
    options_3 = ["A. By dressing up as a mariachi", "B. Singing an emotional song", "C. With a famous mexican dance ", "D. With a magic trick"]
    
    #Define the variable points to calculate the number of correct answers
    points = 0
    
    #Print out the question with respective options using \n format to show options as a list
    print(f"{quest_1} \n {options_1[0]}\n {options_1[1]}\n {options_1[2]}\n {options_1[3]}")
    
    #Store the input of the player as ans_quest_1 
    ans_quest_1 = input(prompt = "Enter your answer: ").upper()
    
    #Define a conditional statement to print out a specific message depending on the input provided by the player
    if ans_quest_1 == "A":
        print("--" * 55)
        print("That is correct! Hector is a an unkown dead willing to do anything to cross over to the world of the living and if he helps Miguel, he can do it")
        print("--" * 55)
        points += 1
    elif ans_quest_1 == "B": 
        print("--" * 55)
        print("Nice try! But Dante is from the land of the living as well as Miguel")
        print("--" * 55)
        
    elif ans_quest_1 == "C": 
        print("--" * 55)
        print("Wrong answer, mama Imelda is from Miguel's family and she doesn't want him to become a musician either.")
        print("--" * 55)
        
    else:
        print("--" * 55)
        print("Nice try, but Chicarrón is going to disappear forever from the Land of the Dead because he does not have anyone remembering him in the living world, he is in the final death")
        print("--" * 55)
     
    #Repeat the process for question 2 
    print(f"{quest_2} \n {options_2[0]}\n {options_2[1]}\n {options_2[2]}\n {options_2[3]}")
    
    ans_quest_2 = input(prompt = "Enter your answer: ").upper()
    
    if ans_quest_2 == "C":
        print("--" * 55)
        print("That is correct! Ernesto de la Cruz is in a particular part of the world of the dead, named the City of the Dead, where an elegant party is held in memory of Ernesto")
        print("--" * 55)
        points += 1
    elif ans_quest_2 == "A": 
        print("--" * 55)
        print("Wrong answer, Ernesto de la Cruz never wanted to visit his living relatives")
        print("--" * 55)
    elif ans_quest_2 == "B": 
        print("--" * 55)
        print("Wrong answer, nice try!")
        print("--" * 55)
    else: 
        print("--" * 55)
        print("Wrong answer, Ernesto de la Cruz is not in the bar")
        print("--" * 55)
    
    #Repeat the process for question 3
    print(f"{quest_3} \n {options_3[0]}\n {options_3[1]}\n {options_3[2]}\n {options_3[3]}")
    
    ans_quest_3 = input(prompt = "Enter your answer: ").upper()
    
    if ans_quest_3 == "B":
        print("--" * 55)
        print("That is correct! In the movie, Miguel sings one of the most famous songs of Ernesto de la Cruz")
        print("--" * 55)
        points += 1
    elif ans_quest_3 == "A": 
        print("--" * 55)
        print("Wrong answer")
        print("--" * 55)
    elif ans_quest_3 == "C": 
        print("--" * 55)
        print("Wrong answer")
        print("--" * 55)
    elif ans_quest_3 == "D": 
        print("--" * 55)
        print("Wrong answer")
        print("--" * 55)
        
    input(prompt = "Press any key to continue")
    
    print("--" * 55)
    
    #Evaluate the points and print out the next part of the story
    if points >= 2:
        print("Good job! You passed level 2 and Miguel is now with his idol, Ernesto de la Cruz, but the story is not what it appears to be, once Héctor guided Miguel to meet Ernesto de la Cruz, he found out that he is not the father of Mama Coco, and their past is correlated with a tragic story that Ernesto had been hiding for a long time, when they were alive, Ernesto and Hector were great musicians and friends who played together, and Hector was the composer of the songs that made Ernesto famous, but Ernesto decided to poison Hector to get a total appropriation of his songs and increase his fame and popularity, something that was never known until Miguel arrived to the land of the dead, so now, you need to defeat Ernesto de la Cruz and help Miguel get back to the land of the living in the next level")
        input(prompt = "Press any key to go to level 3")
        stage_3()
        
    else:
        print("Nice try! but don't worry! You can continue playing but the story is not what it appears to be, in the film, once Héctor guided Miguel to meet Ernesto de la Cruz, he found out that he is not the father of Mama Coco, and their past is correlated with a tragic story that Ernesto had been hiding for a long time, when they were alive, Ernesto and Hector were great musicians and friends who played together, and Hector was the composer of the songs that made Ernesto famous, but Ernesto decided to poison Hector to get a total appropriation of his songs and increase his fame and popularity, something that was never known until Miguel arrived to the land of the dead, so now, you need to defeat Ernesto de la Cruz and help Miguel get back to the land of the living in the next level")
        input(prompt = "Press any key to start level 3")
        stage_3()
        print("--" * 55)
        
    print("--" * 55)

#start defining a def function named stage_3() to store level 3  
def stage_3():
    
    import random
    
    print("--" * 55)
    
    #Provide the instructions to the player
    print("Welcome to the last level, to defeat Ernesto de la Cruz and help Miguel go back to the Land of the living, you must find the hidden word by inserting one letter at a time, could you guess it? Let's find out")
    
    print("--" * 55)
    
    input(prompt = "Press any key to start")
    
    print("--" * 55)
    
    #Define the word in a list
    words = ["aztec", "calavera", "marigold"]
    
    #Define a random.choice in the variable word to select a random word from the variable word
    word = random.choice(words)
    
    #To show the amount of "--" depending on the quantity of letters the word contains
    letters_guessed = "_" * len(word)
    
    #To provide the meaning or description of the word
    hints = ["Ancient civilization that started Día de los Muertos tradition", "Classic flower used to decorate ofrendas guiding departed souls the way to the altar", "Word skull in spanish, symbolizes the sweetness of life"] 
    
    #Define the variable chances 
    chances = 0
    
    #Insert a conditional statement to print out the meaning of the word depending on the random word
    if word == "aztec":
        print("Ancient civilization that started Día de los Muertos tradition")
                
    elif word == "calavera":
        print("Word skull in spanish, is a classic item during Día de los Muertos and symbolizes the sweetness of life")
                
    else:
        print("Colorful flower used to decorate ofrendas guiding departed souls the way to the altar")
    
    #Print out the number of letters the random word contains 
    print(f"This word contains {len(word)} letters")
    
    print("--" * 55)
    
    #Defining a while loop to run the game until the player loses or wins 
    while True:
        
        
        #Print out the letters guessed and characters left
        print(letters_guessed)
        
        try:
            #Ask the player to insert a letter
            letter = input(prompt = "Type any letter: ")
            
            #Conditional statement to evaluate if the input is inside the random word
            if letter in word:
                
                #Insert a for loop to look for the amount of letters that the random word contains
                for i in range(len(word)):
                    
                    #A conditional statement to compare the position of the letter "i" with the input from the player
                    if word[i] == letter:
                        
                        #If the letter guessed is the same as the letter evaluated in position "i" of the random word, the variable is updated
                        letters_guessed = letters_guessed[:i] + letter + letters_guessed[i + 1:]

    
            
            else:
                #When the input is not in the random word, increase by one the variable chance
                chances += 1
            #For the first incorrect answer, print out the head of the hangman
            if chances == 1:
                print("O")
            #For the second incorrect answer, print out the arm of the hangman...
            elif chances == 2:
                print(" O")
                print("/")
            elif chances == 3:
                print(" O")
                print("/|")
            elif chances == 4:
                print(" O")
                print("/|\\")
            elif chances == 5:
                print(" O")
                print("/|\\")
                print("/")
                print("This is your last life")
                
            #If the player fails 6 times, print out the hidden word and break the loop
            elif chances == 6:
                print(" O")
                print("/|\\")
                print("/ \\")
                print(f"You don't have more lifes, the hidden word is {word} ")
                break
                
                
            #If the player guesses all the letters, break the loop and print out message         
            if letters_guessed == word:
                print(f"Great job! the hidden word is {word}")
                break
                
        #Define an exception of errors when inserting values different to text    
        except ValueError:
            print("That is not text, enter only one letter")
            
#Call the first function to run all the code       
stage_1()


print("Excelent job! You finished the game!")
    
end_time = time.time()

#Get the total time in seconds
elapsed_time_seconds = end_time - start_time 

elapsed_time_minutes = elapsed_time_seconds / 60

#Print 
#print(f"Time elapsed: {round(elapsed_time)} seconds")

if elapsed_time_seconds <= 180:
    print(f"That was amazing! {player_name} You finished the game in {elapsed_time_minutes: .2f} minutes , you won!")

else:
    print(f"Nice try! {player_name} You ran out of time to win, you finished in {elapsed_time_minutes: .2f} minutes, but I hope you enjoyed the game")
    
 
print("THANKS FOR PLAYING")

print("SEIZE YOUR MOMENT")
        
    
#helpful references: https://youtu.be/zehwgTB0vV8?si=Zb-Wf50LW79k9TB8, https://youtu.be/EC6PlPUwC5E?si=42dkF2pzRvkgC9Dn
        
        
```
